```python
#INDEXING


import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s)
```

```python
#INDEXING A PARTICULAR VALUE

import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s[3])
```

```python
#INDEXING MULTIPLE VALUES

import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s[[3,5]])
```

```python
#SLICING

import pandas as pd

import numpy as np

data = {1:'shahla',2:'fathm',3:'nsrn'}

s = pd.Series(data,index = [5,1,8,2,3,4])

print(s[0:3])
```

```python
import pandas as pd
import numpy as np

data1 = {'Name':['fathima','nasrin','ke'],'english':[40,60,40],'maths':[40,60,40],'chem':[40,60,40]}
df = pd.DataFrame(data1)

df['social'] = [60,48,50]
df['total marks'] = df['english'] + df['maths'] + df['chem'] + df['social']

list1 = []

for i in df['total marks']:

  if i<=100:
     list1.append('below avrg')
  elif i<=200:
    list1.append('avg')
  else :
    list1.append('excellent')

df['status'] = list1
del df['maths']
df = df.drop('chem',axis = 1) # want to specify axis
df.pop('english')
print('\n')
print(df.columns)
print('\n')
print(df.ndim)
print('\n')
print(df.size)
print('\n')
print(df.shape)

print(df['status'].unique())
print(df['social'].unique())
print(df['status'].value_counts())


print(df)
```
